## Ridge Regularization(L2)
* 1st: Data Preparation.
      - Like making a dataset.
      - from sklearn.datasets import make_regression
* 2nd: Data Scale features (Highly recommended for Ridge).
      - from sklearn.preprocessing import StandardScaler
* 3rd: Initialize and fit the model
      - from sklearn.linear_model import Ridge
* 4th: Make Prediction.
   

```python
from sklearn.datasets import make_regression

X, y = make_regression(
    n_samples = 10000,
    n_features = 40,
    n_targets = 2,
    noise = 0.2,
    n_informative = 15,
    random_state = 43
)
```

```python
import matplotlib.pyplot as plt

# for i in range(X.shape[1]):
#     plt.scatter(X[:, i], y[:, 0])
#     plt.xlabel(f'{i+1}th-Features')
#     plt.ylabel('y-Axis')
#     plt.title(f'{i+1}th Feature and Y plot')
#     plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### 2nd: Scaling Dataset.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scal = scaler.fit_transform(X_train)
X_test_scal = scaler.transform(X_test)
```

##### Now let's see the data distribution.

```python
# for i in range(X.shape[1]):
#     plt.title(f'{i+1}th features vs Y')
#     plt.scatter(X_train_scal[:, i], y_train[:, 0])
#     plt.show()
```

#### 3rd and 4th: Fit and Predict the model.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Fit the model.
lr.fit(X_train_scal, y_train)

# Predict
y_pred = lr.predict(X_test_scal)

# Calculate Accuracy, R2 Score
from sklearn.metrics import mean_squared_error, r2_score
print(f"MSE: {mean_squared_error(y_test, y_pred)}")
print(f"R2-score: ", r2_score(y_pred, y_test))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.001)
```

```python
ridge.fit(X_train_scal, y_train)

# Predict:
y_pred = ridge.predict(X_test_scal)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_pred, y_test))
print("R2-Score: ", r2_score(y_test,y_pred))
```

### Standard method to Find best Alpha

```python
from sklearn.linear_model import RidgeCV
import numpy as np

alphas = np.logspace(-3, 3, 50)  # 0.001 → 1000

ridge_cv = RidgeCV(alphas=alphas, scoring='neg_mean_squared_error', cv=5)
ridge_cv.fit(X_train_scal, y_train)

print("Best alpha:", ridge_cv.alpha_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param_grid = {'alpha': np.logspace(-3, 3, 50)}

grid = GridSearchCV(Ridge(), param_grid, cv=5, scoring='neg_mean_squared_error')
grid.fit(X_train_scal, y_train)

print("Best alpha:", grid.best_params_)
```
