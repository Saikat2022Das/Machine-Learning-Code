## Lasso Regularization
* 1st: Make a Dataset and Split them train and test.
* 2nd: Scaling Them.
* 3rd: Fit and Predict the output.

### 1st: Makeing Dataset and Train test Split

```python
from sklearn.datasets import make_regression

X, y = make_regression(
    n_samples=1000,
    n_features = 23,
    n_targets=1,
    n_informative=12,
    noise = 0.37
)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.23, random_state=42)
```

### 2nd: Make Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit And Transform
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

### 3dr: Fit And Predict the Models

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01)

# Fit and Train
lasso.fit(X_train_scaler, y_train)

# Predict
y_pred = lasso.predict(X_test_scaler)
```

#### Calculate Accuarcy

```python
from sklearn.metrics import r2_score, mean_squared_error

print("MSE: ", mean_squared_error(y_pred, y_test))
print("R2-Score: ", r2_score(y_pred, y_test))
```
