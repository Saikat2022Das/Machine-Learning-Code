```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=23)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
import numpy as np
```

```python
class BatchGDRegressor:

  def __init__(self, learning_rate=0.01, epochs=100):
    self.lr = learning_rate
    self.epochs = epochs
    self.intercept_ = None
    self.coef_ = None

  def fit(self, X_train, y_train):
    # Here We have to update coefficients,

    # 1st: Took one 10 size array, which define as intialize my all coefficients.
    self.intercept_ = 0
    self.coef_ = np.ones(X_train.shape[1])

    for i in range(self.epochs):
      # Update all

      # Update Coefficient
      # b1 = b1 - lr*slop w.r.t b1
      # slop = (-2/n) * All(y_train_i - (mx + b)) * x1
      # 1st Calculate: y-het:
      y_hat = np.dot(X_train, self.coef_) + self.intercept_
      # Dieffernec = y_train_i - (mx + b)
      slop = (-2/X_train.shape[0]) * np.dot((y_train - y_hat), X_train)
      # Update Coefficients
      self.coef_ = self.coef_ - self.lr * slop

      # Update Intercept
      # bo = bo - lr*slop w.r.t bo
      # slop = (-2/n) * All(y_train_i - y_het)

      # We already calculate y-het
      # slop = (-2/X_train.shape[0]) * np.sum(y_train - y_hat)
      slop = -2 * np.mean(y_train - y_hat)  # Both gives same results.
      self.intercept_ = self.intercept_ - self.lr * slop

    print(self.intercept_, self.coef_)

  def predict(self, X_test):
    return np.dot(X_test, self.coef_) + self.intercept_
```

```python
bgd = BatchGDRegressor(learning_rate=0.08, epochs=300)
```

```python
bgd.fit(X_train, y_train)
```

```python
y_pred_class = bgd.predict(X_test)
```

```python
r2_score(y_test, y_pred_class)
```
